## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

```python
colours= np.random.rand(50)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('this is title')
plt.scatter(x,y,c=colours)
plt.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

## The subplot() function in Matplotlib is used to create multiple plots in a single figure. This function allows you to divide the figure into a grid and place a plot in each grid cell, making it easier to compare multiple plots side-by-side or in a specific arrangement. It is especially useful for creating complex visualizations that require the display of multiple related plots.

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))


axs[0, 0].plot(x, y1, 'r')
axs[0, 0].set_title('Line 1: y = 100x')


axs[0, 1].plot(x, y2, 'g')
axs[0, 1].set_title('Line 2: Oscillating Values')


axs[1, 0].plot(x, y3, 'b')
axs[1, 0].set_title('Line 3: y = 10x')

axs[1, 1].plot(x, y4, 'y')
axs[1, 1].set_title('Line 4: Random Variations')


plt.tight_layout()


plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

## A bar plot, or bar chart, is a type of graph that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents. Bar plots are useful for comparing the quantities of different categories, making it easy to visualize differences, trends, and patterns among groups.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize=(10, 6))
plt.barh(company,profit, color='lightgreen')
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

## What is a Box Plot?
A box plot, also known as a whisker plot, is a standardized way of displaying the distribution of data based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It can also show outliers and is used to detect skewness and identify any potential anomalies in the data.

## Why is it Used?
Visualizing Distribution: Box plots provide a clear summary of the distribution of data, highlighting the central tendency, variability, and symmetry.
Identifying Outliers: Box plots are effective at identifying outliers within a dataset.
Comparing Distributions: They are useful for comparing the distributions of multiple datasets.
Detecting Skewness: They help in identifying whether the data is skewed or symmetric.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
boxs= [box1,box2]
```

```python
plt.boxplot(boxs, labels=['Data1', 'Data2'])
plt.show()
```
